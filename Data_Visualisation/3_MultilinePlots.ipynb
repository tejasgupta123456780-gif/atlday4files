## Multiline Plots 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(10)
plt.plot(x, x**2)
plt.plot(x, x**3)
plt.plot(x, x*2)
plt.plot(x, 2**x)
plt.show()
```

```python
plt.plot(x, x**2, x, x**3, x, x*2, x, 2**x)
plt.show()
```

```python
x = np.array([[1, 2, 6, 3], [4, 5, 3, 2]])
plt.plot(x)
plt.show()
```

```python
data = np.random.randn(2, 10)
print(data)
plt.plot(data[0], data[1])
plt.show()
```
